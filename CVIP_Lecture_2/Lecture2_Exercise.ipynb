This is how they did so.

```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
from IPython.display import Image
```

```python
# Read in image to be masked
img_background_bgr = cv2.imread("4.png")
img_background_rgb = cv2.cvtColor(img_background_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=[30,10])
plt.imshow(img_background_bgr)
```

```python
#fetch masking images
mask1 = cv2.imread("10.png", cv2.IMREAD_GRAYSCALE)
mask2 = cv2.imread("11.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=[20,5])
plt.subplot(121);plt.imshow(mask1,cmap='gray')
plt.subplot(122);plt.imshow(mask2,cmap='gray')
```

```python
#combine the masks into one using Bitwise AND
complete_mask = cv2.bitwise_and(mask1, mask2, mask = None)
plt.imshow(complete_mask,cmap='gray')
```

```python
# Apply mask to background image
img_masked = cv2.bitwise_and(img_background_rgb, img_background_rgb, mask=complete_mask)
plt.figure(figsize=[30,10])
plt.imshow(img_masked, cmap = 'gray')
```
